```python
import pandas as pd
import sklearn
import numpy as np
from matplotlib import pyplot as plt
from sklearn.utils import shuffle

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', None)

def print_value(val):
    if val == 1: return "POSITIVE"
    if val == -1: return "NEGATIVE"
    if val == 0: return "NEUTRAL"
```

```python
#Mostra tweets originali
twitter_dataframe = pd.read_csv('dataframes/tweets.csv')
all_user_list = list(np.unique(twitter_dataframe["User"])) #lista di utenti
print("loading CSV: "+str(twitter_dataframe.shape[0])+" tweet found [",len(all_user_list),"users]")
shuffle(twitter_dataframe)
```

```python
#Mostra tweets (con traduzione)
twitter_dataframe = pd.read_csv('dataframes/tweets_translated.csv')
all_user_list = list(np.unique(twitter_dataframe["User"])) #lista di utenti
print("loading CSV: "+str(twitter_dataframe.shape[0])+" tweet found [",len(all_user_list),"users]")
shuffle(twitter_dataframe)
```

```python
#Mostra tweets pre-processati
twitter_dataframe = pd.read_csv('dataframes/pre_processed_tweets.csv')
all_user_list = list(np.unique(twitter_dataframe["User"])) #lista di utenti
print("loading CSV: "+str(twitter_dataframe.shape[0])+" tweet found [",len(all_user_list),"users]")
shuffle(twitter_dataframe)
```

```python
#Mostra la media dei sentimenti:

sentences = pd.read_excel('excel_results/sentences_vader.xlsx',index_col=0)

#Plot a histogram of frequencies
df = pd.DataFrame(sentences["Sentiment"].apply(print_value)).value_counts().plot(kind='barh')

plt.title('Average sentiment of all sentences')
plt.xlabel('Sentences')
plt.savefig('histograms/sentences_sentiment.png', bbox_inches = 'tight')
plt.show(df)
```
